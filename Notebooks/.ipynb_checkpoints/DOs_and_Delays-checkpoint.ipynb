# DO tracings

## Setup

```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
from scipy.ndimage.interpolation import shift
import matplotlib

matplotlib.rcParams['figure.figsize'] = [40, 20]
```

## Delay filter

```python
t = np.linspace(0,10,10*1000)
x = np.exp(-0.1*t) * (np.sin(2 * np.pi * 4 * t) + 0.2*np.sin(2 * np.pi * 8 * t) + 0.02 * np.sin(2 * np.pi * 16 * t))

fs = 1000

x_d = np.copy(x)

x_d = shift(x_d, 1100, cval=0)

x_tot = x + x_d
plt.figure()
plt.subplot(121)
plt.plot(t,x)
plt.plot(t,x_d)
plt.plot(t,x_tot)

plt.xlim((0,10))
plt.subplot(122)

f, Pxx_den = sig.welch(x, fs, nperseg=1024)
f, Pxx_tot = sig.welch(x_tot,fs,nperseg=1024)

plt.plot(f,10*np.log10(Pxx_den))
plt.plot(f,10*np.log10(Pxx_tot),color='green')
plt.xlim((0,30))
```
